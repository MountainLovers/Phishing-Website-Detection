```python
import pandas as pd
```

## 数据载入

```python
legitimate_urls = pd.read_csv("./extracted_csv_files/legitimate-urls.csv")
phishing_urls = pd.read_csv("./extracted_csv_files/phishing-urls.csv")
```

#### 导入待检测url文件

```python
suspicious_urls = pd.read_csv("suspicious-urls.csv")
```

## 数据处理
#### 把`phishing_urls.csv`和`legitimate_urls.csv`两个文件合并作为一个数据集
注：两个文件的列数及内容相同

```python
urls = legitimate_urls.append(phishing_urls)
```

```python
urls.head(5)
```

```python
urls.columns
```

#### 删除掉无用的列（0 - Domain，3-Path，5- Protocol）

```python
urls = urls.drop(urls.columns[[0,3,5]],axis=1)
```

```python
urls.head(5)
```

#### suspicious_urls也删掉无用列

```python
suspicious_urls = suspicious_urls.drop(suspicious_urls.columns[[0,3,5]],axis=1)
```

```python
suspicious_urls.head()
```

#### 由于我们把两个数据集合并为一个数据集，因此前1000行是合法的urls，后1000行是钓鱼urls。因此，如果我们现在拆分数据并为其创建模型将过度拟合，那么在将数据拆分为训练集和测试集之前，我们需要重新整理行，用reset_index打乱顺序

```python
# shuffling the rows in the dataset so that when splitting the train and test set are equally distributed
urls = urls.sample(frac=1).reset_index(drop=True)
```

#### 删除urls数据集中的label这一列

```python
urls_without_labels = urls.drop('label',axis=1)
```

```python
urls_without_labels.columns
```

```python
labels = urls['label']
```

#### 把urls数据集按test_size的比例分割为训练集和测试集

```python
from sklearn.model_selection import train_test_split
```

#### 对计算成功率

```python
data_train, data_test, labels_train, labels_test = train_test_split(urls_without_labels, labels, test_size=0.1, random_state=100)
```

```python
#data_train = urls
#data_test = suspicious_urls
#labels_train = labels
#labels_test = pd.Series([])
```

```python
print(len(data_train),len(data_test),len(labels_train),len(labels_test))
```

```python
labels_train.value_counts()

print("num0: ", end="")
print(labels_train[labels_train == 0].count())
print("num1: ", end="")
print(labels_train[labels_train == 1].count())
```

```python
labels_test.value_counts()
```

## 训练模型

#### 创建模型并将数据拟合到模型中

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(data_train,labels_train)
```

#### 预测测试集的结果（label）

```python
pred_label = model.predict(data_test)
```

```python
print(pred_label)
print(list(labels_test))
```

#### 构造混淆矩阵并检查准确性

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(labels_test,pred_label)
print(cm)
accuracy_score(labels_test,pred_label)
```
